### Introduction
Air Quality Index (AQI) is used to measure the level of air pollution. A **Simple Reflex Agent**
makes decisions based only on the current percept (sensor reading) without considering past data.

In this notebook:
- Sensor input = PM2.5 pollution value
- Agent rule = Determine AQI category
- Output = Air quality condition

The agent reacts immediately using predefined rules.



## AQI Categories (Simplified)

| PM2.5 Value | AQI Category |
|--------------|--------------|
| 0 – 50 | Good |
| 51 – 100 | Moderate |
| 101 – 200 | Unhealthy |
| 201+ | Hazardous |


```

# Simple Reflex Agent for AQI

def get_aqi_category(pm25):
    
    if pm25 <= 50:
        return "Good"
    
    elif pm25 <= 100:
        return "Moderate"
    
    elif pm25 <= 200:
        return "Unhealthy"
    
    else:
        return "Hazardous"


def reflex_agent(pm25_value):
    
    category = get_aqi_category(pm25_value)
    
    print("PM2.5 Value:", pm25_value)
    print("AQI Category:", category)
```

```

# Example sensor readings

values = [35, 80, 150, 250]

for v in values:
    reflex_agent(v)
    print()
```

```

import matplotlib.pyplot as plt

pm_values = [35, 80, 150, 250]

categories = [get_aqi_category(v) for v in pm_values]

plt.figure()
plt.plot(pm_values)
plt.title("PM2.5 Sensor Readings")
plt.xlabel("Observation")
plt.ylabel("PM2.5 Value")
plt.show()
```


## Simple Reflex Agent Structure

Sensor → Agent → Action

Example:

PM2.5 Sensor Reading  
↓  
Reflex Agent (Rule-based system)  
↓  
AQI Category Output



## Conclusion

The Simple Reflex Agent determines the AQI category using predefined rules based on PM2.5 sensor input.
The agent reacts only to the current percept and does not store previous states. This demonstrates
the working principle of simple reflex agents in Artificial Intelligence.
