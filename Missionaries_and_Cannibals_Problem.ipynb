
# Missionaries and Cannibals Problem
## Artificial Intelligence Assignment

### Problem Statement
Three missionaries and three cannibals must cross a river using a boat that can carry at most two people.

### Constraints:
- The boat cannot cross the river empty.
- Cannibals must never outnumber missionaries on either bank (if missionaries are present).
- Goal: Move everyone safely to the right bank.

### State Representation:
(M_left, C_left, Boat_position)

- M_left → Missionaries on left bank
- C_left → Cannibals on left bank
- Boat_position → 1 (Left), 0 (Right)

Initial State: (3, 3, 1)  
Goal State: (0, 0, 0)



## Breadth First Search (BFS)

BFS explores the state space level by level using a Queue (FIFO).
It guarantees the shortest path in an unweighted problem.


```

from collections import deque

# Check valid state
def is_valid(M, C):
    if M < 0 or C < 0 or M > 3 or C > 3:
        return False
    if (M > 0 and C > M):
        return False
    if ((3 - M) > 0 and (3 - C) > (3 - M)):
        return False
    return True

# BFS Implementation
def bfs():
    start = (3, 3, 1)
    goal = (0, 0, 0)
    queue = deque([(start, [])])
    visited = set()

    while queue:
        (M, C, boat), path = queue.popleft()

        if (M, C, boat) == goal:
            return path + [(M, C, boat)]

        if (M, C, boat) in visited:
            continue

        visited.add((M, C, boat))

        moves = [(1,0),(2,0),(0,1),(0,2),(1,1)]

        for m, c in moves:
            if boat == 1:
                new_state = (M - m, C - c, 0)
            else:
                new_state = (M + m, C + c, 1)

            if is_valid(new_state[0], new_state[1]):
                queue.append((new_state, path + [(M, C, boat)]))

    return None

# DFS Implementation
def dfs(state, path, visited):
    M, C, boat = state
    goal = (0, 0, 0)

    if state == goal:
        return path + [state]

    visited.add(state)
    moves = [(1,0),(2,0),(0,1),(0,2),(1,1)]

    for m, c in moves:
        if boat == 1:
            new_state = (M - m, C - c, 0)
        else:
            new_state = (M + m, C + c, 1)

        if is_valid(new_state[0], new_state[1]) and new_state not in visited:
            result = dfs(new_state, path + [state], visited)
            if result:
                return result

    return None

# Run BFS
print("BFS Solution Path:")
solution_bfs = bfs()
for step in solution_bfs:
    print(step)

# Run DFS
print("\nDFS Solution Path:")
solution_dfs = dfs((3,3,1), [], set())
for step in solution_dfs:
    print(step)
```


## Depth First Search (DFS)

DFS explores as deep as possible before backtracking.
It uses recursion (stack).
DFS does not guarantee shortest path.



## Performance Comparison

| Feature | BFS | DFS |
|----------|------|------|
| Data Structure | Queue | Stack |
| Complete | Yes | Not Always |
| Optimal | Yes | No |
| Memory Usage | High | Low |

### Time Complexity:
- BFS: O(b^d)
- DFS: O(b^m)
